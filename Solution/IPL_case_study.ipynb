```python
# Importing

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ball_by = "/content/IPL_Ball_by_Ball_2008_2022.csv"
ball_by = pd.read_csv(ball_by)
ball_by.head(3)
```

```python
matches = "/content/IPL_Matches_2008_2022.csv"
matches = pd.read_csv(matches)
matches.head(10)
```

```python
# Checking shape
print("ball_by_ball shape",ball_by.shape)
print("matches shape",matches.shape)
```

```python
# Checcking dtypes
# ball_by_ball dtypes

ball_by.dtypes
```

```python
# matches dtypes

matches.dtypes
```

```python
# Changing the dtype of matche's date  column

matches['Date'] = pd.to_datetime(matches['Date'],format='%m/%d/%Y',errors='coerce')
matches.dtypes
```

```python
# Checking null values
# ball_by_ball

ball_by.isna().sum()
```

```python
# matches

matches.isnull().sum()
```

```python
# checking duplicates

matches[matches.duplicated()]
```

```python

ball_by[ball_by.duplicated()]
```

```python
# mergining the both ball_by_ball and matches

merged_df = pd.merge(matches,ball_by, on = 'ID')
merged_df.info()
```

```python
# 1. From overs 0-19, which over has the most runs scored in, on average?

ball_by.groupby('overs')['total_run'].mean().idxmax()
```

```python
# 2. Top 10 umpires, who were part of the most matches

matches['Umpire1'].append(matches['Umpire2']).value_counts().nlargest(10)
```

```python
#3. Which team has the highest probability of winning a match given that they had won the toss?

toss_and_won = matches[matches['TossWinner'] == matches['WinningTeam']]
winning_probability = (toss_and_won['TossWinner'].value_counts()/matches['TossWinner'].value_counts()).nlargest(1)
winning_probability
```

```python
#4. Top 10 most successful bowlers in terms of number of wickets taken

ball_by[ball_by['isWicketDelivery'] == 1]['bowler'].value_counts().nlargest(10)
```

```python
#5. Calculate for each team the number of playoff matches played.

playoff = matches[matches['MatchNumber'].isin(['Qualifier 2','Qualifier 1','Qualifier','3rd Place Play-Off'])]
teams_playoff = playoff['Team1'].append(playoff['Team2'])
teams_playoff.value_counts()
```

```python
# 6. On which venue, the team batting second has the most advantage?

# found that Team1 has done batting first

matches[matches['WinningTeam'] == matches['Team2']]['Venue'].value_counts().idxmax()
```

```python
# 7. Which batsman in the history of IPL has been caught out the most?

ball_by[ball_by['kind'] == 'caught'].groupby('player_out')['kind'].value_counts().nlargest(1)
```

```python
# 8. Which bowler has taken the wicket of the batsman in que no. 7?

ball_by[ball_by['player_out'] == 'RG Sharma'].groupby('bowler')['isWicketDelivery'].size().idxmax()
```
